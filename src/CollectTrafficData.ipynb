## Test Travel Times

https://developers.google.com/maps/documentation/directions/intro#traffic-model

https://developers.google.com/maps/documentation/distance-matrix/start

```python
# Driving time (driving) between Vincent's home (486495) and his office (609930)

url = 'https://maps.googleapis.com/maps/api/distancematrix/json?origins=486495+Singapore&destinations=609930+Singapore&key=' + config.gapikey
r = requests.get(url)
data = json.loads(r.text)

# travel time in seconds
data['rows'][0]['elements'][0]['duration']['value']
```

```python
# Transit (train) time between Vincent's home (486495) and his office (609930)

url = 'https://maps.googleapis.com/maps/api/distancematrix/json?origins=486495+Singapore&destinations=609930+Singapore&mode=transit&transit_mode=train&key=' + config.gapikey
r = requests.get(url)
data = json.loads(r.text)

# travel time in seconds
data['rows'][0]['elements'][0]['duration']['value']
```
